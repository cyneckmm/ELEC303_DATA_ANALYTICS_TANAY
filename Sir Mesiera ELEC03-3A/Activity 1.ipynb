<h3>08. Is the a correlation between penguins bill length and bill depth?</h3>

```python
bill_cor
```

<h3>09. Is there a correlation between the species sex and flipper length?</h3>

```python
penguins[penguins['sex'] == 'male']['flipper_length_mm'].sum()
```

```python
penguins[penguins['sex'] == 'female']['flipper_length_mm'].sum()
```

```python
sex_flipperL = f_oneway(penguins[penguins['sex'] == 'male']['flipper_length_mm'],penguins[penguins['sex'] == 'female']['flipper_length_mm'])
```

```python
sex_flipperL
```

<h3>Insight: <strong>There is</strong> a correlation in flipper length to both sexes; however, if we focus on which sex shows a stronger correlation with flipper length, the data indicates that <strong>males have the higher correlation</strong>.</h3>

<h3>10. Is there a correlation between the mass kg that Chinstrap and Gentoo has?</h3>

```python
f_oneway(penguins[penguins['species'] == 'Gentoo']['mass_kg'],penguins[penguins['species'] == 'Chinstrap']['mass_kg'])
```

<h3>Insight: There is a correlation between Chinstrap and Gentoo penguins in terms of body mass (kg).</h3>

<h3>11. Would there be a correlation between all the species in terms of bill depth?</h3>

```python
f_oneway(penguins[penguins['species'] == 'Adelie']['bill_depth_mm'],penguins[penguins['species'] == 'Chinstrap']['bill_depth_mm'])
```

```python
f_oneway(penguins[penguins['species'] == 'Gentoo']['bill_depth_mm'],penguins[penguins['species'] == 'Chinstrap']['bill_depth_mm'])
```

```python
f_oneway(penguins[penguins['species'] == 'Gento']['bill_depth_mm'],penguins[penguins['species'] == 'Adelie']['bill_depth_mm'])
```

<h3>Insight: The data indicates a strong correlation among all species in terms of bill depth</h3>

<h3>12. Is there a correlation between the Adelie and Chinstrap species in terms of bill length (mm)?</h3>

```python
f_oneway(penguins[penguins['species'] == 'Adelie']['bill_length_mm'],penguins[penguins['species'] == 'Chinstrap']['bill_length_mm'])
```

<h3>Insight: Chinstrap and Adelie penguins show a clear correlation in bill length</h3>
